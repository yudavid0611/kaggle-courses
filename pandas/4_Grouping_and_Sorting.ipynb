# 4. Grouping and Sorting

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# sample df
columns = list(load_iris().feature_names)
columns = list(map(lambda x: x[:-5].replace(' ', '_'), columns))

df = pd.DataFrame(columns = columns, data = load_iris().data)
df.head(10)
```

## groupby
- 'DataFrameGroupBy object'를 반환한다.
- 칼럼(들)을 인자로 넣는다. 그러면 칼럼(들)의 unique values을 기준으로 그룹핑이 된다. 

```python
# petal width 각각의 값이 몇 개가 존재하는가?
# 방법1
df.groupby('petal_width').petal_width.size()
```

```python
# petal width 각각의 값이 몇 개가 존재하는가?
# 방법2
df.groupby('petal_width').petal_width.count()
```

```python
# petal_width별 sepal_width의 최대값은 무엇인가?
df.groupby('petal_width').sepal_width.max()
```

```python
# petal_width별 sepal_length의 최소값과 최대값은 무엇인가?
df.groupby('petal_width').sepal_length.agg([min, max])
```

```python
# 위에서 생성한 데이터프레임을 min -> max 칼럼 순으로 내림차순 정렬한다면?
temp = df.groupby('petal_width').sepal_length.agg([min, max])
temp.sort_values(by = ['min', 'max'], ascending=False)
```

```python
# 인덱스를 기준으로 내림차순 정렬하기
df = df.sort_values('sepal_length')     # sepal_length를 기준으로 내림차순 정렬
df = df.sort_index()
df
```
